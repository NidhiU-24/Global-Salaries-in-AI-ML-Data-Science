# **Descriptive Analysis**

"Descriptive Analysis involves summarizing and describing the main features of the dataset. It focuses on organizing and presenting the data in a meaningful way, often using measures such as mean, median, mode, and standard deviation. It provides an overview of the data and helps identify patterns or trends."

```python
# 1. Basic information about the dataset
print(data1.info())
```

```python
# 2. summary statistics for numerical columns
print(data1.describe())
```

```python
# 3. Average salary by employment type
average_salary_by_employment = data1.groupby('employment_type')['salary'].mean().sort_values()
average_salary_by_employment.plot(kind='bar')
plt.title('Average Salary by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Average Salary')
plt.show()
```

# **Exploratory Analysis**
"Exploratory Analysis focuses on exploring and understanding the data without preconceived hypotheses. It involves visualizations, summary statistics, and data profiling techniques to uncover patterns, relationships, and interesting features. It helps generate hypotheses for further analysis"

```python
# 1. Distribution of experience levels
plt.figure(figsize=(10, 6))
sns.countplot(x='experience_level', data=data1, palette='viridis')
plt.title('Distribution of Experience Levels')
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Salary Distribution
plt.figure(figsize=(14, 8))
sns.boxplot(x='job_title', y='salary', data=data1)
plt.title('Salary Distribution by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Salary')
##plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# 3. Distribution of Experience Levels
experience_counts = data1['experience_level'].value_counts()


plt.figure(figsize=(8, 8))
plt.pie(experience_counts, labels=experience_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99','#ffcc99','#ff6666'])
plt.title('Distribution of Experience Levels')
plt.show()
```

The overall amount of work done remotely, possible values are as follows:
1. 0- No remote work (less than 20%)
2. 50- Partially remote/hybird
3. 100 - Fully remote (more than 80%

```python
# 4. Bar chart for remote ratio
plt.figure(figsize=(10, 6))
sns.countplot(x='remote_ratio', data=data1, palette='bright', edgecolor='0.2')
plt.title('Distribution of Remote Work Ratios')
plt.xlabel('Remote Work Ratio')
plt.ylabel('Count')
plt.show()
```

```python
# 5. Company size Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='company_size', data=data1, palette='plasma')
plt.title('Distribution of Company Sizes')
plt.xlabel('Company Size')
plt.ylabel('Count')
plt.show()
```

# **Diagnostic Analysis**
"Diagnostic analysis aims to understand the cause-and-effect relationships within the data. It investigates the factors or variables that contribute to specific outcomes or behaviors. Techniques such as regression analysis or correlation analysis are commonly used."

```python
# 1. Correlation
correlation_matrix = data1.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# 2. Scatter plot with regression line for salary vs. remote work ratio
plt.figure(figsize=(12, 6))
sns.regplot(x='remote_ratio', y='salary', data=data1, scatter_kws={'s': 30}, line_kws={'color': 'red'})
plt.title('Salary vs. Remote Work Ratio')
plt.xlabel('Remote Work Ratio')
plt.ylabel('Salary')
plt.show()
```

 ### This visual explains how the 'remote_ratio' variable is related to the 'salary' variable. The red regression line provides an estimate of the linear relationship between these two variables. The scatter points show individual data points for each observation

```python
# 3. Distribution of skills
skills_columns = ['Python', 'SQL', 'Tableau', 'AWS', 'PyTorch']

fig, axes = plt.subplots(1, 5, figsize=(20, 4))
fig.suptitle('Distribution of Skills', fontsize=16)


axes = axes.flatten()

for i, skill in enumerate(skills_columns):
    skill_counts = data1[skill].value_counts()
    axes[i].pie(skill_counts, labels=skill_counts.index, autopct='%1.1f%%', startangle=90)
    axes[i].set_title(f'Distribution of {skill} Skills')

plt.show()
```

```python
# 4. Comparision of Skills Across Experience Levels
skills_columns = ['Python', 'SQL', 'Tableau', 'AWS', 'PyTorch']


fig, axes = plt.subplots(1, 5, figsize=(20, 6), sharey=True)
fig.suptitle('Comparison of Skills Across Experience Levels', fontsize=16)

for i, skill in enumerate(skills_columns):
    sns.countplot(y='experience_level', hue=skill, data=data1, ax=axes[i], palette='Spectral', edgecolor='0.2')
    axes[i].set_title(f'{skill} Skills')
    axes[i].set_xlabel('Count')
    axes[i].set_ylabel('Experience Level')


plt.tight_layout(rect=[0, 0, 1, 0.9])
plt.show()
```
